```python
# 기초 통계 계산
# lifeExp 열을 연도별로 그룹화 하고 평균을 계산
# 데이터를 year로 그룹화
# liefExp 열의 평균을 계산

group_df = df.groupby('year')
print(group_df)

group_year_lifeExp = group_df['lifeExp']
print(group_year_lifeExp)
```

```python
#시리즈 응용
# 보통 추출할 데이터의 정확한 인덱스를 모르는 경우 추출하는 방법
# boolean 을 통해서 특정조건을 만족하는 값만 출력

scientists = pd.read_csv('scientists.csv')
ages = scientists['Age']
print(ages.mean())
```

```python
# 평균 나이보다 나이가 많은 사람들만 데이터를 추출
ages[ages>ages.mean()]
```

```python
# ages>ages.mean() 의 결과가 여러개가 나오는 것을 볼 수 있다.
# 시리즈나 데이터프레임에 있는 모든 데이터에 대한 한번에 연산하는 것을
# 브로드캐스팅  Broadcasting 이라고 한다
# 그리고 시리즈처럼 여러개의 값을 가닞 데이터는 벡터라고 하고
# 단순 크기를 나타내는 데이터는 스칼라 라고 한다,.
```

```python
print(ages+ages)
# 같은길이의 벡터 값만 계산이 가능하다
```

```python
# 벡터에 스칼라를 연산하여 브로드캐스팅한 결과이다
print(ages+100)
print(ages*100)
```

```python
# 길이가 서로 다른 벡터의 연산
# 시리즈와 시리즈는 같은 인덱스의 값만 계산한다.
# 나저미는 계산이 불가능 하기 때문에 NaN 누락값 으로 처리된다

ages+pd.Series([1,100])
```

```python
# 데이터 프레임 boolean 추출
scientists=scientists[scientists["Age"]>scientists["Age"].mean()]
```

```python
# index가 False인 2행은 출력되지 않는다.
scientists.loc[[True,True,False,True]]
```

```python
# 브로드캐스팅
# 데이터 프레임에 스칼라 연산 적용
scientists*2
# 문자열은 2배로 늘어나고 숫자는 2를 곱한 숫자가 된다.
```

```python
# 시리즈와 데이터 프레임의 형변환과 새로운 열 추가하기
print(scientists['Born'].dtype)
```
